### Embeddings with HuggingFace local model

Import Libraries

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

Create Embedding Model

```python
embedding = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

Text Embeddings

```python
text = 'Mumbai is capital of India'

vector = embedding.embed_query(text)

print(str(vector))
```

Documents Embedding

```python
document = [
    "Delhi is capital of India",
    "Kolkata is capital of west Bengal",
    "Tiger is national aninal of India"
]

doc_vector = embedding.embed_documents(document)

print(str(doc_vector))
```
